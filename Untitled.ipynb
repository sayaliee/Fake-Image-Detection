```python
#importing the required libraries
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import os
import cv2
import tensorflow as tf
from tensorflow.keras import optimizers
from tensorflow.keras.layers import *
from tensorflow.keras.models import Model, load_model
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint
from tensorflow.keras import layers, models, callbacks
```

```python
from tensorflow.keras.layers import BatchNormalization
```

```python
# Global Variables for using them later

IMAGE_SIZE = 128
BATCH_SIZE = 32
CHANNELS = 3
```

```python
# function that would read an image provided the image path, preprocess and return it back
def read_and_preprocess(img_path):
    img = cv2.imread(img_path, cv2.IMREAD_COLOR) # reading the image
    img = cv2.resize(img, (IMAGE_SIZE, IMAGE_SIZE)) # resizing it to 128*128 as given global variable
    img = np.array(img, dtype='float32') # convert its datatype so that it could be normalized
    img = img/255 # normalization (now every pixel is in the range of 0 and 1)
    return img
```

```python
labels = ['real', 'fake'] #let real = 0, fake = 1

X = [] # empty list To store images
y = [] # empty list To store labels

data_path = './dataset/' # path containing data and image samples
```

```python
for folder in os.scandir(data_path):
    for entry in os.scandir(data_path + folder.name):

        X.append(read_and_preprocess(data_path + folder.name + '/' + entry.name))
        
        if folder.name[0]=='r':
            y.append(0) # real
        else:
            y.append(1) # fake
```

```python
#since X array stores the images, will find the shape of it
X = np.array(X)
X.shape
```

We have 2041 images with height and width of 128 and 3 color channels 

```python
y = np.array(y) #Y stored the labels if the img is real or fake
y.shape
```

```python
# plotting random images

import random

plt.figure(figsize=(16, 16))

for i in range(12):
    ax = plt.subplot(3, 4, i + 1)
    plt.axis("off")
    img = random.randint(0, 1289)
    plt.imshow(X[img])
    plt.title(labels[y[img]])
```

```python
#Now, we will split the data for training, validation and testing process
#Training = 80%, Validation = 10%, Testing = 10%
from sklearn.model_selection import train_test_split
```

```python
X_train, X_val, y_train, y_val = train_test_split(X, y, 
                                                  test_size=0.2, 
                                                  shuffle=True, 
                                                  stratify=y, 
                                                  random_state=123)
#To maintain the ratio of each class (stratify= True) 
#To get same samples every time we split our data (random state)
# this will split the data into 0.8(training) and 0.2(testing+validation)
```

```python
X_test, X_val, y_test, y_val = train_test_split(X_val, y_val, 
                                                  test_size=0.5, 
                                                  shuffle=True, 
                                                  stratify=y_val, 
                                                  random_state=123)
#now we will again split the data in half , so 0.5 for each validation and testing
```

```python
input_shape = (IMAGE_SIZE, IMAGE_SIZE, CHANNELS)
n_classes = 2
```

Now that our data is divided properly for training, validation and testing, we will now build a sequential model using tensorflow and keras

```python
model2 = tf.keras.Sequential([
    Conv2D(32, (3,3), activation='relu', input_shape=(128, 128, 3)),
    MaxPooling2D(2, 2),
    Dropout(0.2),

    Conv2D(64, (3,3), activation='relu'),
    MaxPooling2D(2,2),
    Dropout(0.3),

    Flatten(),
    Dense(128, activation='relu'),
    Dropout(0.5),
    Dense(1, activation='sigmoid')
])
```

```python
# There are 7.3 Million Traininable Parameters
model2.summary()
```

```python
# compile the model
model2.compile(loss = 'binary_crossentropy', optimizer='adam', metrics= ["accuracy"])
```

```python
# use early stopping to exit training if validation loss is not decreasing even after certain epochs (patience)
earlystopping = EarlyStopping(monitor='val_accuracy', mode='max', verbose=1, patience=7)
```

```python
# save the best model with most accuracy
#The ModelCheckpoint callback is used to save the model during training
checkpointer = ModelCheckpoint(filepath="fakevsreal_weights.keras", verbose=1,mode='min', save_best_only=True)
```

```python
history = model2.fit(X_train, y_train, epochs = 100, validation_data=(X_val, y_val), batch_size=BATCH_SIZE, shuffle=True, callbacks=[earlystopping, checkpointer])
```
